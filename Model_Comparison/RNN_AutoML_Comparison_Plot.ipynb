```
import pandas as pd
automl = pd.read_csv('0214_true_vs_pred_automl.csv')
rnn = pd.read_csv('0214_true_vs_pred_rnn.csv')
```

```
automl = automl.iloc[4:]
automl.head()
```

```
rnn.head()
```

```
trueClose = automl['True Close'].tolist()
automlClose = automl['Predicted Close'].tolist()
rnnClose = rnn['Predicted Close'].tolist()
```

```
%matplotlib inline
import time
import pylab as pl
from IPython import display
from matplotlib import pyplot as plt
```

```
x = list(range(len(rnn)))
plt.plot(x,trueClose,'-b',label='True')
plt.plot(x,rnnClose,'-r', label='RNN')
plt.plot(x,automlClose,'-g', label='AutoML')
plt.xlim(0, 100)
plt.ylim(322, 326.5)
plt.legend()
plt.show()
```
