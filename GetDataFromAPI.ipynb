```
# Get latest 5 days of data
url = 'https://intraday.worldtradingdata.com/api/v1/intraday?symbol=AAPL&interval=1&range=5&api_token=ixRdaL4qjumQzCDpEdsQvrsqNmGUDlJ83KhJRNDX5pxU7Ddzbk7AguOvVdnz'
response = requests.get(url)
out = response.json()
```

```
data = take(len(out['intraday']),out['intraday'].items())
dic = {'Time stamp':[],'Open':[],'Close':[],'High':[],'Low':[],'Volume':[]}
for item in data:
    dic['Time stamp'].append(item[0])
    dic['Open'].append(float(item[1]['open']))
    dic['Close'].append(float(item[1]['close']))
    dic['High'].append(float(item[1]['high']))
    dic['Low'].append(float(item[1]['low']))
    dic['Volume'].append(float(item[1]['volume']))

import pandas as pd
df = pd.DataFrame.from_dict(dic)

df.to_csv('apple_5days_data.csv',index=None, header=True)
```
